```python
import requests
import numpy as np
import pandas as pd
import time
import re
import datetime
import selenium
from selenium import webdriver as wd 
from selenium.webdriver import Chrome
import urllib.request
from bs4 import BeautifulSoup
```

```python
url = 'https://www.youtube.com/feed/trending'
req = requests.get(url).text
req
```

```python
soup = BeautifulSoup(req,'html.parser')
print(soup.prettify())
```

```python
soup_find = soup.findAll('h3',{'class':'yt-lockup-title'})
soup_find
```

```python
soup_img = soup.findAll('img')[8:]
soup_img
```

```python
title = []
total_time = []
url = []
Thumbnail = []

for i in range(len(soup_find)):
    a = soup_find[i].text.split(' - 길이: ')
    # 영상 제목 크롤링
    title.append(a[0])
    
    # 영상 url 크롤링 
    url.append('https://www.youtube.com'+soup_find[i].a.get('href'))
    
    # 영상 시간 크롤링
    a = a[1].split(':')
    if len(a) == 3:
        second = int(a[0])*3600 + int(a[1])*60 + int(a[2])
        
    elif len(a) == 2:
        second = int(a[0])*60 + int(a[1])

    else:
        second = int(a[0])
    total_time.append(second)
    
    #####################################################
    # 썸네일 사진 URL 
    a = soup_img[i].get('src').split('?sqp')[0]
    if a[:4] == 'http':
        Thumbnail.append(a)
    else:
        Thumbnail.append(soup_img[i].get('data-thumb').split('?sqp')[0])
    
df = pd.DataFrame({
    'title':title,
    'total_time':total_time,
    'url':url,
    'Thumbnail':Thumbnail
})
df
```
